```python
import pandas as pd

items_df=pd.read_csv("D:/☆디지털혁신팀/☆입찰 단가 분석/☆데이터/확정내역_신정뉴타운.csv",encoding="utf-8")
codes_df=pd.read_csv("D:/☆디지털혁신팀/☆입찰 단가 분석/☆데이터/CBS ITEM 조회_20200414.csv",encoding="utf-8")
```

```python
items_df=items_df.fillna(" ")
items_df["item"]=items_df["공종명"]+" "+items_df["규격"]
items_df=items_df.loc[220:320]
items_df.index=range(len(items_df))
```

```python
codes_df=codes_df.fillna(" ")
codes_df["code"]=codes_df["명칭"]+" "+codes_df["규격"]
codes_df.index=range(len(codes_df))
```

```python
from konlpy.tag import Okt

okt=Okt()

stop_words=["(",")"]

morphs=[]
item_tokens=[]
for sentence in items_df["item"]:
    new_word=[]
    morphs=okt.morphs(sentence,stem=False)
    for word in morphs:
        if word not in stop_words:
            new_word.append(word)
    item_tokens.append(new_word)
            
morphs=[]
code_tokens=[]
for sentence in codes_df["code"]:
    new_word=[]
    morphs=okt.morphs(sentence,stem=False)
    for word in morphs:
        if word not in stop_words:
            new_word.append(word)
    code_tokens.append(new_word)
```

```python
import numpy as np

def Jaccard_similarity(doc1,doc2):
    doc1=set(doc1)
    doc2=set(doc2)
    return (len(doc1&doc2)/len(doc1|doc2)+len(doc1&doc2)/len(doc1)+len(doc1&doc2)/len(doc2))/3

for i in items_df.index:
    item=items_df.loc[i,"item"]
    recomm1_sim=3e-30
    recomm2_sim=2e-30
    recomm3_sim=1e-30
    recomm1="No Match"
    recomm2="No Match"
    recomm3="No Match"
    recomm1_cbs="No Match"
    recomm2_cbs="No Match"
    recomm3_cbs="No Match"
    for j in codes_df.index:
        try:
            sim=Jaccard_similarity(item_tokens[i],code_tokens[j])
            code=codes_df.loc[j,"code"]
            cbs=codes_df.loc[j,"CBS코드"]
            if sim>recomm1_sim:
                recomm3_sim=recomm2_sim
                recomm2_sim=recomm1_sim
                recomm1_sim=sim
                recomm3=recomm2
                recomm2=recomm1
                recomm1=code
                recomm3_cbs=recomm2_cbs
                recomm2_cbs=recomm1_cbs
                recomm1_cbs=cbs
            elif recomm1_sim>sim>recomm2_sim:
                recomm3_sim=recomm2_sim
                recomm2_sim=sim
                recomm3=recomm2
                recomm2=code
                recomm3_cbs=recomm2_cbs
                recomm2_cbs=cbs
            elif recomm2_sim>sim>recomm3_sim:
                recomm3_sim=sim
                recomm3=code
                recomm3_cbs=cbs
        except:
            continue
    items_df.loc[i,"추천1"]=recomm1
    items_df.loc[i,"추천1_cbs"]=recomm1_cbs
    items_df.loc[i,"추천2"]=recomm2
    items_df.loc[i,"추천2_cbs"]=recomm2_cbs
    items_df.loc[i,"추천3"]=recomm3
    items_df.loc[i,"추천3_cbs"]=recomm3_cbs
    try:
        items_df.to_csv("D:/☆디지털혁신팀/☆입찰 단가 분석/☆데이터/확정내역_신정뉴타운_맵핑 3.csv",encoding="euc-kr")
    except:
        continue
```
