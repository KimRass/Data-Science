# English

```python
urllib.request.urlretrieve("https://wit3.fbk.eu/get.php?path=XML_releases/xml/ted_en-20160408.zip&filename=ted_en-20160408.zip", filename="ted_en-20160408.zip")

#xml 파일로부터 <content>와 </content> 사이의 내용만 가져온다.
with zipfile.ZipFile("ted_en-20160408.zip", "r") as z:
    ted = etree.parse(z.open("ted_en-20160408.xml", "r"))
    ted = "\n".join(ted.xpath("//content/text()"))
```

```python
ted[:300]
```

```python
# 정규 표현식의 sub 모듈을 통해 content 중간에 등장하는 (Audio), (Laughter) 등의 배경음 부분을 제거.
ted = re.sub(r"\([^)]*\)", "", ted)

# 입력 코퍼스에 대해서 NLTK를 이용하여 문장 토큰화를 수행.
sents = nltk.tokenize.sent_tokenize(ted)

sents_norm = []
#각 문장에 대해서 구두점을 제거하고, 대문자를 소문자로 변환합니다.
for sent in tqdm(sents):
    sent = re.sub(r"[^a-z0-9]+", " ", sent.lower())
    sents_norm.append(sent)
```

```python
#각 문장에 대해서 nltk.tokenize.word_tokenize()를 수행합니다.
path = "ted_en_nltk.json"
if os.path.exists(path):
    with open(path, "r", encoding="utf-8") as f:
        corpus = json.load(f)
else:
    corpus = [nltk.tokenize.word_tokenize(sent) for sent in tqdm(sents_norm)]
    with open(path, "w", encoding="utf-8") as f:
        json.dump(corpus, f, ensure_ascii=False, indent="\t")
```

```python
print("총 샘플의 개수 : {}".format(len(corpus)))
```

```python
for line in corpus[:3]:
    print(line)
```

```python
%%time
if os.path.exists("ted_en_w2v"):
    #모델을 불러옵니다.
    model = gensim.models.KeyedVectors.load_word2vec_format("ted_en_w2v")
else:
    #모델을 생성합니다.
    model = gensim.models.Word2Vec(corpus, size=100, window=5, min_count=5, workers=4, sg=0)
    #모델을 저장합니다.
    model.wv.save_word2vec_format("ted_en_w2v")
```

- `size` : 임베딩 벡터의 차원.
- `min_count` : 단어 최소 빈도 수(빈도가 적은 단어들은 학습하지 않는다)
- `workers` : 학습을 위한 프로세스 수  
- `sg=0` : cBoW
- `sg=1` : Skip-gram.  

```python
model.wv.most_similar("man")
```

# Korean Language

```python
import string
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings.txt", filename="ratings.txt")
train_data = pd.read_table("ratings.txt")
train_data = train_data[["document", "label"]]
train_data = train_data.dropna()

#한글이 아닌 문자를 제거합니다.
train_data["document"] = train_data["document"].str.replace(r"[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

#불용어를 정의합니다.
sw = set()
for punct in string.punctuation:
    sw.add(punct)
# sw = ["의","가","이","은","들","는","좀","잘","걍","과","도","를","으로","자","에","와","한","하다"]
```

```python
train_data.head()
```

```python
path = "ratings_train_okt.json"
if os.path.exists(path):
    with open(path, "r", encoding="utf-8") as f:
        train_tkn = json.load(f)
else:
    train_tkn = []
    for doc in tqdm(train_data["document"]):
        tkn = okt.morphs(doc, stem=True, norm=True)
        tkn = [word for word in tkn if not word in sw]
        train_tkn.append(tkn)
    
    with open(path, "w", encoding="utf-8") as f:
        json.dump(train_tkn, f, ensure_ascii=False, indent="\t")
```

```python
print(f"리뷰의 최대 길이 : {max(len(doc) for doc in train_tkn)}")
print(f"리뷰의 평균 길이 : {np.mean(list(map(len, train_tkn)))}")
```

```python
train_tkn
```

```python
fig, ax = plt.subplots()

ax.hist([len(doc) for doc in train_tkn], bins=50)

fig.tight_layout()
```

```python
%%time
path = "ratings_train_w2v"
if os.path.exists(path):
    #모델을 불러옵니다.
    model = gensim.models.KeyedVectors.load_word2vec_format(path)
else:
    #모델을 생성합니다.
    model = gensim.models.Word2Vec(train_tkn, size=100, window=5, min_count=5, workers=4, sg=0)
    #모델을 저장합니다.
    model.wv.save_word2vec_format(path)
```

```python
#완성된 임베딩 매트릭스의 크기를 확인합니다.
model.wv.vectors.shape
```

```python
print(model.wv.most_similar("안성기"))
print(model.wv.most_similar("히어로"))
print(model.wv.most_similar("미국"))
```
